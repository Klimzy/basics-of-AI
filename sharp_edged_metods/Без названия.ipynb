```python
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
```

```python
def scobellFilter(img):
    scobelMask = np.array([[1, 0, -1],
                           [1, 0, -1],
                           [1, 0, -1]])
    
    height = img.shape[0]
    wight = img.shape[1]
    scobelImage = np.zeros_like(img)
    
    for i in range(1, height-1):
        for j in range(1, wight-1):
            sum =  img[i-1][j-1] * scobelMask[0][0] + img[i-1][j] * scobelMask[0][1] + img[i-1][j+1] * scobelMask[0][2] 
            sum += img[i][j-1] * scobelMask[1][0] + img[i][j] * scobelMask[1][1] + img[i][j+1] * scobelMask[1][2]
            sum += img[i+1][j-1] * scobelMask[2][0] + img[i+1][j] * scobelMask[2][1] + img[i+1][j+1] * scobelMask[2][2]
            scobelImage[i][j] = sum
            
    return scobelImage
            
            
            
```
